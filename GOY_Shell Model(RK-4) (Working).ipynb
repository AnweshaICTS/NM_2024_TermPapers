```python
#Since, we pass an array it will also get modified with every itteration Mind it!!!

@njit
def iterator(Tot_step,y,T): 
    
    for i in range(1,Tot_step):
        
        ka= np.zeros(N+1,dtype = 'complex128')
        kb= np.zeros(N+1,dtype = 'complex128')

        kc= np.zeros(N+1,dtype = 'complex128')
        kd= np.zeros(N+1,dtype = 'complex128')
        
        ka,kb,kc,kd = K(T[i-1],y[:,i-1],step)
        
        y[:,i] = y[:,i-1] + (step/6)*(ka+2*kb+2*kc+kd) 
```

```python
%%time

lamb = 2
nu = 1e-7
epsilon = 0.5 


# Number of variables
N = 22

k_0 = 1/16 + 0J
a = 1
b =  -epsilon
c = -1 + epsilon
k_n = np.append(k_0,k_0*(lamb**np.arange(1,N+1,dtype = "float64"))) # n = 0,1,2,3,...N


# Force term Final L

force = 5*(1+1J)*10e-4

F = np.zeros(N+1,dtype = 'c16')

# F[4] = F[4] + force


#Time 
step = 1e-4

final_time = 20

tot_step = int(final_time/step)

t = np.linspace(0,final_time,num = tot_step,endpoint= True)

total_repeat = 20
cores = 4

each_realization = int((total_repeat/cores))
```

```python
# #initialize the itterator
Y = np.column_stack(( np.append(np.sqrt(k_n[0:3])*np.exp(1J*np.random.uniform(0,np.pi,3)),np.sqrt(k_n[3:])*np.exp(-1*k_n[3:]**2)*np.exp(1J*np.random.uniform(0,np.pi,N-2))),
                           np.zeros([N+1,tot_step-1],dtype = 'c16')))

iterator(1,Y,t)
```

```python
def repeator(repeat,seeding): 
    
    np.random.seed(seeding)
    

    vel_profile_1 = np.zeros((4,N+1),dtype = 'float64')

    
    struct_fn_2 = np.zeros((4, N+1))

    
    
    for _ in range(repeat):
        
    # inital condition
    
        init = np.append(np.sqrt(k_n[0:3])*np.exp(1J*np.random.uniform(0,2*np.pi,3),dtype='complex128'),np.sqrt(k_n[3:])*np.exp(-1*k_n[3:]**2,dtype= 'complex128')*np.exp(1J*np.random.uniform(0,2*np.pi,N-2)))
        
        Y = np.column_stack((init,np.zeros([N+1,tot_step-1],dtype = 'complex128')))
        

        iterator(tot_step,Y,t)  #Modifies the Y !!!
        
        for dumm_dumm in range(1,21):
            
        
            vel_profile_1[0,dumm_dumm] = np.abs(np.imag(Y[dumm_dumm + 2,-1]*Y[dumm_dumm+1,-1]*Y[dumm_dumm,-1] - (0.25)*Y[dumm_dumm - 1,-1]*Y[dumm_dumm,-1]*Y[dumm_dumm +1,-1]))**(5/3)
            vel_profile_1[1,dumm_dumm] = np.abs(np.imag(Y[dumm_dumm + 2,-1]*Y[dumm_dumm+1,-1]*Y[dumm_dumm,-1] - (0.25)*Y[dumm_dumm - 1,-1]*Y[dumm_dumm,-1]*Y[dumm_dumm +1,-1]))**(2)
            vel_profile_1[2,dumm_dumm] = np.abs(np.imag(Y[dumm_dumm + 2,-1]*Y[dumm_dumm+1,-1]*Y[dumm_dumm,-1] - (0.25)*Y[dumm_dumm - 1,-1]*Y[dumm_dumm,-1]*Y[dumm_dumm +1,-1]))**(7/3)
            vel_profile_1[3,dumm_dumm] = np.abs(np.imag(Y[dumm_dumm + 2,-1]*Y[dumm_dumm+1,-1]*Y[dumm_dumm,-1] - (0.25)*Y[dumm_dumm - 1,-1]*Y[dumm_dumm,-1]*Y[dumm_dumm +1,-1]))**(8/3)
            
            
        struct_fn_2 =  struct_fn_2 + vel_profile_1
        
        
        
    return struct_fn_2
```

```python
%%time

final_data = np.zeros((4,N+1))

args=[(each_realization,np.random.randint(1,100)),(each_realization,np.random.randint(1,100)),(each_realization,np.random.randint(1,100)),(each_realization,np.random.randint(1,100))]

with concurrent.futures.ProcessPoolExecutor() as executor:
    
    results = [executor.submit(repeator,a,b) for (a,b) in args]
    
    for f in concurrent.futures.as_completed(results):
        
        final_data = final_data + f.result()

final_data = (final_data)/cores

# np.save('GOY_shell_Model_'+str(N)+'_structure_fn_2',final_data)
```
